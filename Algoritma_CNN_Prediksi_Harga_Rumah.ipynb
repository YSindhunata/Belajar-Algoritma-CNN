```python
# Impor library yang diperlukan
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from google.colab import drive

# Mengautentikasi dan meng-mount Google Drive
drive.mount('/content/drive')

# Load dataset (perbarui path sesuai lokasi file di Google Drive)
df = pd.read_csv('/content/drive/My Drive/Data/harga_rumah.csv', delimiter=';')

# Menampilkan nama-nama kolom untuk verifikasi
print("Nama-nama kolom dalam dataset:", df.columns)

# Memastikan nama kolom benar
if 'harga' in df.columns:
    X = df.drop(['id_rumah', 'harga'], axis=1)  # Fitur-fitur
    y = df['harga']  # Target (harga rumah)
else:
    raise ValueError("Kolom 'harga' tidak ditemukan dalam dataset.")

# Normalisasi fitur menggunakan StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Memisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Membangun model Neural Network
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(1)  # Layer output untuk prediksi harga rumah (satu nilai kontinu)
])

# Kompilasi model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Melatih model dengan data latih
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluasi model dengan data uji
loss, mae = model.evaluate(X_test, y_test)

print(f'Loss: {loss}, MAE: {mae}')

# Fungsi untuk prediksi harga rumah berdasarkan input baru
def predict_new_house(luas_rumah, kamar_tidur):
    # Data input baru
    new_data = np.array([[luas_rumah, kamar_tidur]])

    # Normalisasi data input baru menggunakan skala yang sama dengan data latih
    new_data_scaled = scaler.transform(new_data)

    # Prediksi harga menggunakan model yang telah dilatih
    prediction = model.predict(new_data_scaled)

    return prediction[0][0]

# Mendapatkan input dari pengguna
rumah_id = input("Masukkan ID Rumah: ")
luas_rumah = float(input("Masukkan Luas Rumah (m2): "))
kamar_tidur = int(input("Masukkan Jumlah Kamar Tidur: "))

# Prediksi harga berdasarkan input pengguna
predicted_price = predict_new_house(luas_rumah, kamar_tidur)
print(f"Prediksi harga rumah dengan id {rumah_id}, luas {luas_rumah} m2, kamar tidur {kamar_tidur}: {predicted_price}")
```
